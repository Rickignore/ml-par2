# Ensemble Learning

<i>Ensemble Learning</i> adalah teknik dalam <i>Machine Learning</i> yang menggabungkan beberapa model untuk meningkatkan akurasi, stabilitas, dan performa prediksi. Tujuan utama dari <i>Ensemble Learning</i> adalah mengurangi kesalahan (bias dan varians) yang mungkin terjadi jika hanya menggunakan satu model.

Jenis-jenis Ensemble Learning:
- Bagging (Random Forest)
- Boosting
- Stacking
