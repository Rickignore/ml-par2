# Perbedaan Visualisasi Data Iris menggunakan PCA dan LDA

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.datasets import load_iris
```

PCA dan LDA dapat digunakan untuk reduksi dimensi sehingga dapat digunakan untuk plot. Misalnya pada data input terdapat 4 dimensi, dapat direduksi menjadi 2 dimensi sehingga data tersebut dapat di plot menjadi 2 dimensi juga.

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
print(f"Jumlah data sebelum dilakukan reduksi dimensi = {x.shape}")
```

```python
#1. Reduksi dimensi menggunakan PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x)

#Visualisasi PCA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_pca[:,0], y = x_pca[:,1], hue = y, palette = 'Set1', legend='full')
plt.title("PCA -> Reduksi dimensi dari 4D ke 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

#2. Reduksi dimensi menggunakan LDA
lda = LDA(n_components = 2)
x_lda = lda.fit_transform(x,y)

#Visualisasi LDA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_lda[:,0], y = x_lda[:,1], hue = y, palette = 'Set1', legend='full')
plt.title("LDA -> Reduksi dimensi dari 4D ke 2D")
plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.show()
```

```python
import pandas as pd
data_file = os.path.join(dataset_path, 'Students data.csv')
data = pd.read_csv(data_file)
data
```

```python
data.columns
```

```python
x = data[['Algebra', 'Calculus1',
       'Calculus2', 'Statistics', 'Probability', 'Measure',
       'Functional_analysis']] #7D
y = data['y']
```

```python
#1. Reduksi dimensi menggunakan PCA
pca = PCA(n_components = 2)
x_pca = pca.fit_transform(x)

#Visualisasi PCA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_pca[:,0], y = x_pca[:,1], hue = y, palette = 'Set1', legend='full')
plt.title("PCA -> Reduksi dimensi dari 7D ke 2D")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

#2. Reduksi dimensi menggunakan LDA
lda = LDA(n_components = 2)
x_lda = lda.fit_transform(x,y)

#Visualisasi LDA
plt.figure(figsize=(8,6))
sns.scatterplot(x = x_lda[:,0], y = x_lda[:,1], hue = y, palette = 'Set1', legend='full')
plt.title("LDA -> Reduksi dimensi dari 7D ke 2D")
plt.xlabel("Linear Discriminant 1")
plt.ylabel("Linear Discriminant 2")
plt.show()
```

PCA dan LDA belum dapat melakukan visualisasi dengan baik pada data labek 0, 1, dan 2. Namun demikian, dengan melakuka reduksi data 7 dimensi, tetap dapat dilakukan menjadi 2 dimensi dengan menggunakan PCA dan LDA
