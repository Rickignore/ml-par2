```python

#https://www.google.com/url?q=https%3A%2F%2Fwww.kaggle.com%2Fdatasets%2Ffatemehmehrparvar%2Fobesity-levels
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fatemehmehrparvar/obesity-levels")

print("Path to dataset files:", path)
```

```python
import os
dataset_path = '/kaggle/input/obesity-levels'
files = os.listdir(dataset_path)
print("Nama Files: ",files)
```

```python
import pandas as pd
data_file = os.path.join(dataset_path, 'ObesityDataSet_raw_and_data_sinthetic.csv')
data = pd.read_csv(data_file)
data
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Gender_encoded'] = label_encoder.fit_transform(data['Gender'])
data['CALC_encoded'] = label_encoder.fit_transform(data['CALC'])
data['FAVC_encoded'] = label_encoder.fit_transform(data['FAVC'])
data['SCC_encoded'] = label_encoder.fit_transform(data['SCC'])
data['SMOKE_encoded'] = label_encoder.fit_transform(data['SMOKE'])
data['family_history_with_overweight_encoded'] = label_encoder.fit_transform(data['family_history_with_overweight'])
data['CAEC_encoded'] = label_encoder.fit_transform(data['CAEC'])
data['MTRANS_encoded'] = label_encoder.fit_transform(data['MTRANS'])
data['NObeyesdad_encoded'] = label_encoder.fit_transform(data['NObeyesdad'])
data
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data
```

```python
data.info()
```

```python
x = data[['Age','Height','FCVC','NCP','CH2O','FAF','TUE','Gender_encoded','CALC_encoded',
          'FAVC_encoded','SCC_encoded','SMOKE_encoded','family_history_with_overweight_encoded',
          'CAEC_encoded','MTRANS_encoded']]
y = data['Weight']
```


# Ensemble Learning for Regression
## Begging

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)

#Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 0)
rf_model.fit(x_train, y_train)

#Prediksi
y_pred = rf_model.predict(x_test)
#Evaluasi
print(f'MSE = {mean_squared_error(y_pred, y_test)}')
print(f'MAE = {mean_absolute_error(y_pred, y_test)}')
```

# Boosting

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)

#Base model
base_model = DecisionTreeRegressor(max_depth = 10)

#AdaBoost Regressor
ada_model = AdaBoostRegressor(estimator = base_model, n_estimators = 100, random_state = 0)
ada_model.fit(x_train, y_train)

#Prediksi
y_pred = ada_model.predict(x_test)

#Evaluasi
print(f'MSE = {mean_squared_error(y_pred, y_test)}')
print(f'MAE = {mean_absolute_error(y_pred, y_test)}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)

#Gradient Boosting Regressor
gb_model = GradientBoostingRegressor(n_estimators = 100, learning_rate=0.1, max_depth = 7, random_state = 0)
gb_model.fit(x_train, y_train)

#Prediksi
y_pred = gb_model.predict(x_test)

#Evaluasi
print(f'MSE = {mean_squared_error(y_pred, y_test)}')
print(f'MAE = {mean_absolute_error(y_pred, y_test)}')
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state = 0)

#XGB Regressor
xgb_model = xgb.XGBRegressor(n_estimators = 100, learning_rate=0.1, max_depth = 10, random_state = 0)
xgb_model.fit(x_train, y_train)

#Prediksi
y_pred = xgb_model.predict(x_test)

#Evaluasi
print(f'MSE = {mean_squared_error(y_pred, y_test)}')
print(f'MAE = {mean_absolute_error(y_pred, y_test)}')
```

##Stacking

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

base_model = [
    ('rf', RandomForestRegressor(n_estimators=100, random_state=0)),
    ('svr', SVR(kernel='rbf')),
    ('gbr', GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=10, random_state=0))
]

meta_learner = LinearRegression()

stacking_model = StackingRegressor(estimators=base_model, final_estimator=meta_learner)
stacking_model.fit(x_train, y_train)

#prediksi
y_pred = stacking_model.predict(x_test)

print(f"MSE = {mean_squared_error(y_pred, y_test)}")
print(f"MAE = {mean_absolute_error(y_pred, y_test)}")
```

##Setelah menggunakan stacking classifier yang terdiri dari RandomForestRegressor, SVR, dan GradientBoostingRegressor dengan Linear Regression sebagai meta learner, nilai MAE dan MSE justru meningkat.
